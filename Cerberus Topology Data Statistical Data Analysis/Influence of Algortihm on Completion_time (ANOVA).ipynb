# Homogeneity of variance 

To test homogenity I calculate ratio between maximum standard deviation and minimum standard deviaton.
If the ration smaller than 2, then the assumption of homogenity of variance will be fullfilled. 

```python
threshold=2

ratio = sample_df.groupby('algorithm_id').std().max() / sample_df.groupby('algorithm_id').std().min()

if float(ratio) < 2 :
    print("Homogenity of variance condition is satisfied. The ratio is:",ratio)
else:
    print("Homogenity of variance condition is not satisfied. The ratio is:",ratio)


```

The ratio of the largest to the smallest sample standard deviation is 1.96,Thethreshold value is 2,
Thus third assumption is fullfilled. Next step is Hypothesis testing

# Hypothesis Testing
According to five steps process of hypothesis testing:
    
H₀: μ1= μ2 = μ3 = μ4 = μ5 = μ6 = μ7= μ8 = μ9 = μ10  // All sample mean completion time of each algorithm_id's are equal
    
    
H₁: Not all completion_time means are equal
    
    
α = 0.05


According to F test statistics:

# Create ANOVA  table

```python

data = [['Between Groups', '', '', '', '', '', ''], ['Within Groups', '', '', '', '', '', ''], ['Total', '', '', '', '', '', '']] 
anova_table = pd.DataFrame(data, columns = ['Source of Variation', 'SS', 'df', 'MS', 'F', 'P-value', 'F crit']) 
anova_table.set_index('Source of Variation', inplace = True)
```

```python
print(anova_table)
```

# calculate SSTR and update anova table

```python
x_bar = sample_df['completion_time'].mean()
SSTR = sample_df.groupby('algorithm_id').count() * (sample_df.groupby('algorithm_id').mean() - x_bar)**2
anova_table['SS']['Between Groups'] = SSTR['completion_time'].sum()
```

# calculate SSE and update anova table

```python
SSE = (sample_df.groupby('algorithm_id').count() - 1) * sample_df.groupby('algorithm_id').std()**2
anova_table['SS']['Within Groups'] = SSE['completion_time'].sum()
```

```python
anova_table
```

# calculate SSTR and update anova table

```python
SSTR = SSTR['completion_time'].sum() + SSE['completion_time'].sum()
                                           
anova_table['SS']['Total'] = SSTR
```

# update degree of freedom

```python
anova_table['df']['Between Groups'] = sample_df['algorithm_id'].nunique() - 1
anova_table['df']['Within Groups'] = sample_df.shape[0] - sample_df['algorithm_id'].nunique()
anova_table['df']['Total'] = sample_df.shape[0] - 1
```

```python
anova_table
```

# calculate MS

```python
anova_table['MS'] = anova_table['SS'] / anova_table['df']
```

# calculate F 

```python
F = anova_table['MS']['Between Groups'] / anova_table['MS']['Within Groups']
anova_table['F']['Between Groups'] = F
```

# p-value

```python
anova_table['P-value']['Between Groups'] = 1 - stats.f.cdf(F, anova_table['df']['Between Groups'], anova_table['df']['Within Groups'])
```

# F critical 

```python
alpha = 0.05
#H1: μ ≠ μ 0, where a difference is hypothesized and this is called a two-tailed test

tail_hypothesis_type = "two-tailed"
if tail_hypothesis_type == "two-tailed":
    alpha /= 2
anova_table['F crit']['Between Groups'] = stats.f.ppf(1-alpha, anova_table['df']['Between Groups'], anova_table['df']['Within Groups'])
```

```python

anova_table
```

```python
anova_table['P-value']['Between Groups'] < alpha
```

# The p-value approach

```python
print("Approach 1: The p-value approach to hypothesis testing in the decision rule")
conclusion = "Failed to reject the null hypothesis."
if anova_table['P-value']['Between Groups'] <= alpha:
    conclusion = "Null Hypothesis is rejected."
print("F-score is:", anova_table['F']['Between Groups'], " and p value is:", anova_table['P-value']['Between Groups'])    
print(conclusion)
```

Since the p-value PR(>F) is less than our error rate (0.05 > 0.010566), we could reject the null hypothesis. This means we are quite confident that there is a different in completion_time for algorithm_id.

# The critical value approach

```python
print("Approach 2: The critical value approach to hypothesis testing in the decision rule")
conclusion = "Failed to reject the null hypothesis."
if anova_table['F']['Between Groups'] > anova_table['F crit']['Between Groups']:
    conclusion = "Null Hypothesis is rejected."
print("F-score is:", anova_table['F']['Between Groups'], " and critical value is:", anova_table['F crit']['Between Groups'])
print(conclusion)
```

# As a Conclusion it is strong evidence that not all average completion_time are the same for  different algorithm_ids , at 5% significance level. Thus the algorithm_id has no significant influence on completion_time.
